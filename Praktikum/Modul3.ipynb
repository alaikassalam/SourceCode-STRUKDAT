Buat program dengan menggunakan bahasa pemrograman Python untuk soal-soal berikut ini:
(a) Buat fungsi untuk mebuat list baru dari suatu list (parameter dari fungsi), dimana list baru berisi
data-data yang berbeda (tidak duplikat) dari list yang terdapat pada parameter, misalkan:
input atau paramater: [1,3,1,1,4,5,7,7,1,5]
output-return value : [1,3,4,5,7]
(b) Buat fungsi untuk menghasilkan list baru yang berisi karakter-karakter konsonan yang berbeda
dari suatu string yang terdapat pada parameter, misalkan :
input atau parameter :’’struktur data’’
output-return value : [’s’,’t’,’r’,’k’,’d’]

```python
def newlist(data):
    hasil = []
    for index in data:
        if index not in hasil:
            hasil.append(index)
    return hasil
data = [1,3,1,1,4,5,7,7,1,5]
result = newlist(data)
print(f'input atau paramater : {data}')
print(f'output-return value : {result}')
```

```python
def nyarikonsonan(karakter):
    hasil = []
    vokal = ['a','i','u','e','o',' ']
    for index in karakter:
        if index not in vokal and index not in hasil:
            hasil.append(index)
    return hasil
karakter = 'struktur data'
result = nyarikonsonan(karakter)
print(result)
```

```python
def stack():
    s = []
    return s

def push(s, data):
    s.append(data)
    return s

def pop(s):
    data = s.pop()
    return data

def peek(s):
    return s[len(s) - 1]

def isEmpty(s):
    return s == []

def parenthesesCheck(matStr):
    s = stack()
    kurungBuka = "({["
    kurungTutup = ")}]"
    proses = ""
    error = ""
    i = 1
    j = 1
    for ch in matStr:
        if ch in kurungBuka:
            push(s, ch)
            proses += str(f"{i}. Push stack with {ch} --> {s}\n")
            i += 1
        elif ch in kurungTutup:
            top = peek(s)
            if not isEmpty(s):
                pop(s)
                proses += str(f"{i}. Pop stack --> {s}\n")
                i += 1
                if kurungBuka.index(top) != kurungTutup.index(ch):
                    error += str(f"{j}. Kurung buka '{top}' tidak cocok dengan kurung tutup '{ch}'\n")
                    j += 1
    if isEmpty(s) and error == "":
        return True , proses, error
    else:
        return False, proses, error

matStr = "([4+5]/{9+8]*(3+2]}"
[result, proses, error] = parenthesesCheck(matStr)
print(result)
print(f"Proses cek :\n{proses}")
print(f"Error :\n{error}")
```

```python
def stack():
    s = []
    return s

def push(s, data):
    s.append(data)

def pop(s):
    if not isEmpty(s):
        return s.pop()
    else:
        return None

def peek(s):
    if not isEmpty(s):
        return s[-1]
    else:
        return None

def isEmpty(s):
    return s == []

def size(s):
    return len(s)

def evaluatePostfix(expression):
    st = st()
    operators = set(['+', '-', '*', '/'])
    
    for token in expression.split():
        if token not in operators:
            push(st, int(token))
        else:
            operand2 = pop(st)
            operand1 = pop(st)
            if operand1 is None or operand2 is None:
                return "Invalid expression"
            if token == '+':
                push(st, operand1 + operand2)
            elif token == '-':
                push(st, operand1 - operand2)
            elif token == '*':
                push(st, operand1 * operand2)
            elif token == '/':
                push(st, operand1 / operand2)
    
    if size(st) == 1:
        return pop(st)
    else:
        return "Invalid expression"

def displayEvaluationProcess(expression):
    st = stack()
    operators = set(['+', '-', '*', '/'])
    process = []
    
    for token in expression.split():
        if token not in operators:
            push(st, int(token))
        else:
            operand2 = pop(st)
            operand1 = pop(st)
            if operand1 is None or operand2 is None:
                return "Invalid expression", []
            if token == '+':
                result = operand1 + operand2
            elif token == '-':
                result = operand1 - operand2
            elif token == '*':
                result = operand1 * operand2
            elif token == '/':
                result = operand1 / operand2
            push(st, result)
            process.append(f"{operand1} {token} {operand2} = {result}")
    
    if size(st) == 1:
        return pop(st), process
    else:
        return "Invalid expression", []

# Contoh penggunaan
expression = "3 4 + 5 *"
result, process = displayEvaluationProcess(expression)
print("Proses Evaluasi:")
for step, details in enumerate(process, start=1):
    print(f"Langkah {step}: {details}")
print("\nHasil Evaluasi:", result)
```

```python
def stack():
    s = []
    return s

def push(s, data):
    s.append(data)

def pop(s):
    if not isEmpty(s):
        return s.pop()
    else:
        return None

def peek(s):
    if not isEmpty(s):
        return s[-1]
    else:
        return None

def isEmpty(s):
    return s == []

def parenthesesValidation(expression):
    s = stack()
    matching = {')': '(', ']': '[', '}': '{'}
    process = []
    errors = []
    
    for i, ch in enumerate(expression, start=1):
        if ch in '([{':
            push(s, (ch, i))
            process.append(f"{i}. Push stack with '{ch}' --> {s}")
        elif ch in ')]}':
            if isEmpty(s) or matching[ch] != peek(s)[0]:
                errors.append(f"Error at position {i}: Mismatched or unpaired '{ch}'")
            else:
                pop(s)
                process.append(f"{i}. Pop stack --> {s}")
    
    while not isEmpty(s):
        ch, pos = pop(s)
        errors.append(f"Error at position {pos}: No matching closing bracket for '{ch}'")
    
    return len(errors) == 0, process, errors

# Contoh penggunaan
expression = "{[3+4]*5}/(6-2)}"
result, process, errors = parenthesesValidation(expression)
print("Validasi Susunan Tanda Kurung:")
for step, details in enumerate(process, start=1):
    print(f"Langkah {step}: {details}")
if result:
    print("\nEkspresi matematika adalah susunan tanda kurung yang valid.")
else:
    print("\nEkspresi matematika memiliki kesalahan pada susunan tanda kurung:")
    for error in errors:
        print(error)
```

```python
def calculator(expression):
    s = stack()
    operators = {'+', '-', '*', '/', '%'}
    operands_count = 0
    operators_count = 0
    brackets_count = 0
    process = []  # Menyimpan string penjelasan proses perhitungan
    
    for user_input in expression.split():
        if user_input.isdigit():
            push(s, int(user_input))
            operands_count += 1
            process.append(f"Push operand {int(user_input)} ke dalam stack")
        elif user_input in operators:
            if len(s) < 2:
                print("Error: Tidak cukup operand untuk operasi.")
                return None
            operand2 = pop(s)
            operand1 = pop(s)
            if operand1 is None or operand2 is None:
                print("Error: Tidak cukup operand untuk operasi.")
                return None
            operators_count += 1
            
            process.append(f"Pop operand {operand1} dari stack")
            process.append(f"Pop operand {operand2} dari stack")
            
            if user_input == '+':
                result = operand1 + operand2
                process.append(f"Operasi {operand1} + {operand2} = {result}")
            elif user_input == '-':
                result = operand1 - operand2
                process.append(f"Operasi {operand1} - {operand2} = {result}")
            elif user_input == '*':
                result = operand1 * operand2
                process.append(f"Operasi {operand1} * {operand2} = {result}")
            elif user_input == '/':
                if operand2 == 0:
                    print("Error: Pembagian dengan nol.")
                    return None
                result = operand1 / operand2
                process.append(f"Operasi {operand1} / {operand2} = {result}")
            elif user_input == '%':
                if operand2 == 0:
                    print("Error: Modulo dengan nol.")
                    return None
                result = operand1 % operand2
                process.append(f"Operasi {operand1} % {operand2} = {result}")
                
            push(s, result)
            process.append(f"Push hasil operasi {result} ke dalam stack")
        else:
            print("Error: Input tidak valid.")
            return None
            
    if len(s) != 1:
        print("Error: Operasi belum selesai.")
        return None
    elif operands_count - 1 != operators_count or brackets_count != 0:
        print("Error: Ekspresi tidak valid.")
        return None
    else:
        return peek(s), process

# Contoh penggunaan
expression = "3 4 + 5 *"
result, process = calculator(expression)
if result is not None:
    print("Hasil perhitungan:", result)
    print("Proses perhitungan:")
    for step, explanation in enumerate(process, start=1):
        print(f"Langkah {step}: {explanation}")
```

```python
def tangkap_nama(nama_dicari, antrian):
    s = stack()
    putaran = 0
    proses = []  # Menyimpan string penjelasan proses penyelesaian

    while not isEmpty(antrian):
        putaran += 1
        depan_antrian = pop(antrian)
        proses.append(f"Putaran {putaran}: Nama yang ada di depan antrian adalah {depan_antrian}")

        if depan_antrian == nama_dicari:
            proses.append("Nama yang dicari ditemukan di depan antrian. Permainan berhenti.")
            return putaran, proses
        else:
            push(s, depan_antrian)
            proses.append(f"Nama {depan_antrian} tidak sesuai dengan yang dicari. Dipindahkan ke dalam tumpukan.")

        if isEmpty(antrian):
            while not isEmpty(s):
                push(antrian, pop(s))
            proses.append("Antrian telah habis diproses, namun nama yang dicari masih belum ditemukan. Menambahkan kembali nama dari tumpukan ke antrian untuk putaran berikutnya.")

    proses.append("Semua putaran telah dilakukan, namun nama yang dicari tidak ditemukan di dalam antrian.")
    return putaran, proses

# Contoh penggunaan
nama_dicari = "jaki"
antrian = ["ita", "bayu", "jaki", "andi"]
putaran, proses = tangkap_nama(nama_dicari, antrian)
print("Total putaran yang diperlukan:", putaran)
print("Proses penyelesaian:")
for langkah, penjelasan in enumerate(proses, start=1):
    print(f"Langkah {langkah}: {penjelasan}")
```
