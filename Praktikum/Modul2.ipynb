```python
def mul_matrix(matriks1, matriks2):
    perkalian = []  # Membuat list kosong untuk hasil perkalian matriks
    if len(matriks1[0]) == len(matriks2):  # Memeriksa apakah jumlah kolom matriks pertama sama dengan jumlah baris matriks kedua
        for i in range(len(matriks1)):  # Iterasi untuk setiap baris matriks pertama
            perkalian.append([])  # Menambahkan list kosong untuk setiap baris hasil perkalian
            for j in range(len(matriks2[0])):  # Iterasi untuk setiap kolom matriks kedua
                sum = 0  # Variabel untuk menyimpan hasil penjumlahan perkalian elemen
                for k in range(len(matriks1[0])):  # Iterasi untuk mengalikan dan menjumlahkan elemen-elemen matriks
                    sum += matriks1[i][k] * matriks2[k][j]  # Menjumlahkan hasil perkalian elemen
                perkalian[i].append(sum)  # Menambahkan hasil perkalian ke dalam matriks hasil perkalian
        return perkalian  # Mengembalikan hasil perkalian matriks
    else:
        return "error ---require 3 but found 2"  # Mengembalikan pesan kesalahan jika jumlah kolom matriks pertama tidak sama dengan jumlah baris matriks kedua
```
- Fungsi `mul_matrix` menerima dua argumen: `matriks1` dan `matriks2` (matriks yang akan dikalikan).

- Memeriksa apakah jumlah kolom matriks pertama sama dengan jumlah baris matriks kedua, syarat yang diperlukan untuk melakukan perkalian matriks.
- Jika syarat tersebut terpenuhi, melakukan perkalian matriks sesuai aturan perkalian matriks.
- Setiap elemen hasil perkalian dimasukkan ke dalam matriks hasil perkalian.
- Jika syarat tersebut tidak terpenuhi, mengembalikan pesan kesalahan.

```python
def createSparseMatrix(rows, cols):
    matrix = {}  # Membuat matriks sebagai kamus (dictionary)
    elements = int(input("Jumlah elemen : "))  # Meminta jumlah elemen matriks dari pengguna
    matrix["rows"] = rows  # Menambahkan jumlah baris ke matriks
    matrix["cols"] = cols  # Menambahkan jumlah kolom ke matriks
    for i in range(elements):
        row = int(input("Baris ke - ? : "))  # Meminta baris dari pengguna
        col = int(input("Kolom ke - ? : "))  # Meminta kolom dari pengguna
        value = int(input(f"data [{row}][{col}]: "))  # Meminta nilai elemen dari pengguna
        matrix[(row, col)] = value  # Menambahkan nilai elemen ke matriks dengan menggunakan tupel (row, col) sebagai kunci
    return matrix  # Mengembalikan matriks sparse yang sudah terisi
```
- Fungsi ini digunakan untuk membuat matriks sparse dengan jumlah baris dan kolom tertentu.
- Pertama, jumlah elemen matriks diminta dari pengguna.
- Kemudian, baris, kolom, dan nilai dari setiap elemen dimasukkan ke dalam matriks sebagai kamus, menggunakan tupel (row, col) sebagai kunci.
- Matriks sparse yang sudah terisi kemudian dikembalikan.

```python
def addSparseMatrix(mat1, mat2):
    result = {}  # Membuat matriks hasil penjumlahan sebagai kamus (dictionary)
    if mat1["rows"] != mat2["rows"] or mat1["cols"] != mat2["cols"]:  # Memeriksa apakah ukuran kedua matriks sama

    result["rows"] = mat1["rows"]  # Menambahkan jumlah baris matriks hasil ke matriks hasil
    result["cols"] = mat1["cols"]  # Menambahkan jumlah kolom matriks hasil ke matriks hasil

            if (i, j) in mat1:  # Jika elemen ada dalam matriks pertama

            if value != 0:  # Jika nilai hasil penjumlahan bukan 0
```
