```python
# clasc Node
class Node:  # Mendefinisikan sebuah kelas bernama Node
    def __init__(self, item):  # Konstruktor untuk inisialisasi node baru
        self.data = item  # Menyimpan data yang diterima sebagai parameter
        self.next = None  # Menginisialisasi penunjuk next dengan None (belum menunjuk ke node lain)

    def getData(self):  # Mendefinisikan metode untuk mengakses data dari node
        return self.data  # Mengembalikan data yang disimpan dalam node

    def getNext(self):  # Mendefinisikan metode untuk mengakses node berikutnya
        return self.next  # Mengembalikan penunjuk ke node berikutnya

    def setData(self, item):  # Mendefinisikan metode untuk mengubah data dalam node
        self.data = item  # Mengatur data dalam node dengan nilai baru yang diterima sebagai parameter

    def setNext(self, pt):  # Mendefinisikan metode untuk mengubah penunjuk next dalam node
        self.next = pt  # Mengatur penunjuk next dengan node yang diterima sebagai parameter
```

```python
# class LinkedList
class LinkedList:  # Mendefinisikan sebuah kelas bernama LinkedList
    def __init__(self):  # Konstruktor untuk inisialisasi linked list baru
        self.head = None  # Menginisialisasi head (kepala) linked list sebagai None (belum ada node)

    def isEmpty(self):  # Metode untuk memeriksa apakah linked list kosong
        return self.head == None  # Mengembalikan True jika head adalah None, menunjukkan list kosong

    def addNode(self, item):  # Metode untuk menambah node baru di depan linked list
        temp = Node(item)  # Membuat node baru dengan data item
        temp.setNext(self.head)  # Mengatur node baru menunjuk ke head saat ini
        self.head = temp  # Mengatur head menjadi node baru

    def size(self):  # Metode untuk menghitung jumlah node dalam linked list
        current = self.head  # Memulai dari head
        count = 0  # Inisialisasi penghitung
        while current != None:  # Loop hingga akhir linked list
            count += 1  # Increment penghitung untuk setiap node
            current = current.getNext()  # Pindah ke node berikutnya
        return count  # Mengembalikan total jumlah node

    def displayList(self):  # Metode untuk menampilkan data dalam linked list
        current = self.head  # Memulai dari head
        temp = '['  # Inisialisasi string hasil
        while current != None:  # Loop hingga akhir linked list
            temp += str(current.getData()) + ' '  # Tambahkan data node ke string hasil
            current = current.getNext()  # Pindah ke node berikutnya
        temp += ']'  # Tutup string hasil dengan tanda ]
        return temp  # Mengembalikan string hasil

    def search(self, data):  # Metode untuk mencari node dengan data tertentu
        current = self.head  # Memulai dari head
        while current != None:  # Loop hingga akhir linked list
            if data == current.getData():  # Jika data ditemukan
                return True  # Mengembalikan True
            current = current.getNext()  # Pindah ke node berikutnya
        return False  # Mengembalikan False jika data tidak ditemukan

    def addRear(self, newData):  # Metode untuk menambah node baru di belakang linked list
        current = self.head  # Memulai dari head
        temp = Node(newData)  # Membuat node baru dengan data newData
        if not current:  # Jika linked list kosong, head menjadi node baru
            self.head = temp
            return
        while current.getNext() != None:  # Loop hingga node terakhir
            current = current.getNext()  # Pindah ke node berikutnya
        current.setNext(temp)  # Mengatur node terakhir menunjuk ke node baru

    def insertAfter(self, data, target):  # Metode untuk menyisipkan node baru setelah node tertentu
        current = self.head  # Memulai dari head
        temp = Node(data)  # Membuat node baru dengan data
        found = False  # Inisialisasi penanda ditemukan
        while current is not None and not found:  # Loop hingga node ditemukan atau akhir linked list
            if current.getData() == target:  # Jika target ditemukan
                found = True  # Atur penanda ditemukan menjadi True
            else:
                current = current.getNext()  # Pindah ke node berikutnya
        if found:  # Jika node target ditemukan
            temp.setNext(current.getNext())  # Atur node baru menunjuk ke node setelah target
            current.setNext(temp)  # Atur node target menunjuk ke node baru

    def removeFront(self):  # Metode untuk menghapus node di depan linked list
        if self.head is None:  # Jika linked list kosong
            return "List kosong"  # Mengembalikan pesan "List kosong"
        else:
            self.head = self.head.getNext()  # Mengatur head menjadi node berikutnya

    def removeRear(self):  # Metode untuk menghapus node di belakang linked list
        if self.head is None:  # Jika linked list kosong
            return "List kosong"  # Mengembalikan pesan "List kosong"
        elif self.head.getNext() is None:  # Jika hanya ada satu node
            self.head = None  # Mengatur head menjadi None
        else:
            current = self.head  # Memulai dari head
            while current.getNext().getNext() is not None:  # Loop hingga node kedua terakhir
                current = current.getNext()  # Pindah ke node berikutnya
            current.setNext(None)  # Menghapus node terakhir

    def remove(self, data):  # Metode untuk menghapus node dengan data tertentu
        current = self.head  # Memulai dari head
        previous = None  # Inisialisasi node sebelumnya
        found = False  # Inisialisasi penanda ditemukan
        while not found:  # Loop hingga node ditemukan
            if current.getData() == data:  # Jika data ditemukan
                found = True  # Atur penanda ditemukan menjadi True
            else:
                previous = current  # Set previous ke current
                current = current.getNext()  # Pindah ke node berikutnya
        if previous is None:  # Jika node yang dihapus adalah head
            self.head = current.getNext()  # Mengatur head menjadi node berikutnya
        else:
            previous.setNext(current.getNext())  # Mengatur node sebelumnya menunjuk ke node setelah node yang dihapus
```

```python
def remainderFunction(data, num):
    return data % num  # Mengembalikan sisa pembagian (modulus) dari data dengan num

def createHashTable(size):
    hashTable = [[None] for _ in range(size)]  # Membuat tabel hash dengan ukuran size, diinisialisasi dengan daftar yang berisi None
    return hashTable  # Mengembalikan tabel hash yang telah dibuat

def chaining(dataList, hashTable):
    for data in dataList:  # Iterasi melalui setiap elemen dalam dataList
        temp = remainderFunction(data, len(hashTable))  # Menghitung indeks hash menggunakan remainderFunction
        if hashTable[temp][0] is None:  # Jika slot hash kosong (None)
            hashTable[temp] = [data]  # Menempatkan data di slot tersebut sebagai daftar baru
        else:
            hashTable[temp].append(data)  # Jika slot hash tidak kosong, tambahkan data ke dalam daftar di slot tersebut
    return hashTable  # Mengembalikan tabel hash yang telah diisi dengan data

def searchHash(data, hashTable):
    temp = remainderFunction(data, len(hashTable))  # Menghitung indeks hash menggunakan remainderFunction
    if hashTable[temp][0] is None:  # Jika slot hash kosong (None)
        return False  # Data tidak ditemukan, mengembalikan False
    else:
        for i in range(len(hashTable[temp])):  # Iterasi melalui daftar di slot hash
            if hashTable[temp][i] == data:  # Jika data ditemukan
                return f"('data berada di slot ke-', {temp}, 'dan indeks ke-', {i})"  # Mengembalikan posisi data
    return False  # Data tidak ditemukan setelah iterasi, mengembalikan False
```
