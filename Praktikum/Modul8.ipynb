Buatlah program untuk mendapatkan path untuk jalur terpendek dan terpanjang dari sebuah graph. jalur terpendek / terpanjang HARUS lebih dari 1 dan ditampilkan semua. Misalkan terdapat Graph pada Gambar 5

````python
graph = {
  "a": ["b", "c", "d"],
  "b": ["a", "c", "e"],
  "c": ["a", "b", "d", "e"],
  "d": ["a", "c", "e"],
  "e": ["b", "c", "d"],
}

def semuaPath(graph, awal, akhir, path=[]):
  path = path + [awal]
  if awal == akhir:
      return [path]
  elif awal not in graph:
      return []
  hasilPath = []
  for node in graph[awal]:
      if node not in path:
          pathNew = semuaPath(graph, node, akhir, path)
          for pathBaru in pathNew:
              hasilPath.append(pathBaru)
  return hasilPath

def terpendekPath(graph, awal, akhir, path=[]):
  path = path + [awal]
  if awal == akhir:
      return path
  if awal not in graph:
      return None
  terpendek = None
  for node in graph[awal]:
      if node not in path:
          pathBaru = terpendekPath(graph, node, akhir, path)
          if pathBaru:
              if not terpendek or len(pathBaru) < len(terpendek):
                  terpendek = pathBaru
  return terpendek

def terpanjangPath(graph, awal, akhir, path=[]):
  path = path + [awal]
  if awal == akhir:
      return [path]
  if awal not in graph:
      return None
  hasilPath = []
  panjang = 0
  for node in graph[awal]:
      if node not in path:
          pathNew = terpanjangPath(graph, node, akhir, path)
          for pathBaru in pathNew:
              if len(pathBaru) > panjang:
                  hasilPath = [pathBaru]
                  panjang = len(pathBaru)
              elif len(pathBaru) == panjang:
                  hasilPath.append(pathBaru)
  return hasilPath

print("Jalur Terpendek:")
print(terpendekPath(graph, "a", "d"))
print()
print("Jalur Terpanjang:")
print(terpanjangPath(graph, "a", "d"))
print()
print("Semua Jalur:")
print(semuaPath(graph, "a", "d"))
````

````python
graph = {
  "a": ["b", "c", "d"],  # Node "a" memiliki tetangga "b", "c", dan "d"
  "b": ["a", "c", "e"],  # Node "b" memiliki tetangga "a", "c", dan "e"
  "c": ["a", "b", "d", "e"],  # Node "c" memiliki tetangga "a", "b", "d", dan "e"
  "d": ["a", "c", "e"],  # Node "d" memiliki tetangga "a", "c", dan "e"
  "e": ["b", "c", "d"],  # Node "e" memiliki tetangga "b", "c", dan "d"
}

def semuaPath(graph, awal, akhir, path=[]):
  path = path + [awal]  # Menambahkan node awal ke path
  if awal == akhir:  # Jika node awal sama dengan node akhir, return path
      return [path]
  elif awal not in graph:  # Jika node awal tidak ada di graph, return list kosong
      return []
  hasilPath = []  # List untuk menyimpan semua jalur yang ditemukan
  for node in graph[awal]:  # Iterasi melalui semua tetangga dari node awal
      if node not in path:  # Cek apakah node belum ada di path
          pathNew = semuaPath(graph, node, akhir, path)  # Rekursi untuk menemukan semua path dari tetangga
          for pathBaru in pathNew:  # Tambahkan setiap path baru ke hasilPath
              hasilPath.append(pathBaru)
  return hasilPath  # Return semua jalur yang ditemukan

def terpendekPath(graph, awal, akhir, path=[]):
  path = path + [awal]  # Menambahkan node awal ke path
  if awal == akhir:  # Jika node awal sama dengan node akhir, return path
      return path
  if awal not in graph:  # Jika node awal tidak ada di graph, return None
      return None
  terpendek = None  # Variabel untuk menyimpan jalur terpendek
  for node in graph[awal]:  # Iterasi melalui semua tetangga dari node awal
      if node not in path:  # Cek apakah node belum ada di path
          pathBaru = terpendekPath(graph, node, akhir, path)  # Rekursi untuk menemukan jalur terpendek dari tetangga
          if pathBaru:  # Jika pathBaru tidak None
              if not terpendek or len(pathBaru) < len(terpendek):  # Cek apakah pathBaru lebih pendek dari terpendek
                  terpendek = pathBaru
  return terpendek  # Return jalur terpendek

def terpanjangPath(graph, awal, akhir, path=[]):
  path = path + [awal]  # Menambahkan node awal ke path
  if awal == akhir:  # Jika node awal sama dengan node akhir, return path
      return [path]
  if awal not in graph:  # Jika node awal tidak ada di graph, return None
      return None
  hasilPath = []  # List untuk menyimpan jalur terpanjang
  panjang = 0  # Variabel untuk menyimpan panjang jalur terpanjang
  for node in graph[awal]:  # Iterasi melalui semua tetangga dari node awal
      if node not in path:  # Cek apakah node belum ada di path
          pathNew = terpanjangPath(graph, node, akhir, path)  # Rekursi untuk menemukan jalur terpanjang dari tetangga
          for pathBaru in pathNew:  # Cek setiap path baru
              if len(pathBaru) > panjang:  # Jika panjang pathBaru lebih panjang dari panjang, update hasilPath dan panjang
                  hasilPath = [pathBaru]
                  panjang = len(pathBaru)
              elif len(pathBaru) == panjang:  # Jika panjang pathBaru sama dengan panjang, tambahkan pathBaru ke hasilPath
                  hasilPath.append(pathBaru)
  return hasilPath  # Return jalur terpanjang

print("Jalur Terpendek:")
print(terpendekPath(graph, "a", "d"))  # Panggil fungsi terpendekPath untuk menemukan jalur terpendek dari "a" ke "d"
print()
print("Jalur Terpanjang:")
print(terpanjangPath(graph, "a", "d"))  # Panggil fungsi terpanjangPath untuk menemukan jalur terpanjang dari "a" ke "d"
print()
print("Semua Jalur:")
print(semuaPath(graph, "a", "d"))  # Panggil fungsi semuaPath untuk menemukan semua jalur dari "a" ke "d"
````

````python
### Penjelasan Fungsi dan Proses Penyelesaian dengan Contoh

Mari kita jelaskan langkah-langkah dan proses penyelesaian dari setiap fungsi menggunakan contoh graf yang sudah diberikan.

#### 1. Fungsi `semuaPath`

Fungsi ini menemukan semua jalur yang mungkin dari node awal ke node akhir dalam graf.

**Proses:**
- Mulai dari node awal.
- Tambahkan node awal ke path saat ini.
- Jika node awal adalah node akhir, kembalikan path.
- Jika node awal tidak ada di graf, kembalikan list kosong.
- Untuk setiap tetangga node awal, jika tetangga belum ada di path saat ini, rekursi untuk menemukan semua path dari tetangga ke node akhir.
- Gabungkan semua path baru ke dalam hasil path.

**Contoh:**
Untuk mencari semua jalur dari "a" ke "d":

1. Mulai dari "a", path = ["a"]
2. Tetangga "a": ["b", "c", "d"]
3. Pilih "b":
   - Path = ["a", "b"]
   - Tetangga "b": ["a", "c", "e"]
   - Pilih "c":
     - Path = ["a", "b", "c"]
     - Tetangga "c": ["a", "b", "d", "e"]
     - Pilih "d":
       - Path = ["a", "b", "c", "d"]
       - "d" adalah node akhir, simpan path ini.
   - Pilih "e":
     - Path = ["a", "b", "e"]
     - Tetangga "e": ["b", "c", "d"]
     - Pilih "c":
       - Path = ["a", "b", "e", "c"]
       - Tetangga "c": ["a", "b", "d", "e"]
       - Pilih "d":
         - Path = ["a", "b", "e", "c", "d"]
         - "d" adalah node akhir, simpan path ini.
4. Pilih "c":
   - Path = ["a", "c"]
   - Tetangga "c": ["a", "b", "d", "e"]
   - Pilih "d":
     - Path = ["a", "c", "d"]
     - "d" adalah node akhir, simpan path ini.
   - Pilih "e":
     - Path = ["a", "c", "e"]
     - Tetangga "e": ["b", "c", "d"]
     - Pilih "d":
       - Path = ["a", "c", "e", "d"]
       - "d" adalah node akhir, simpan path ini.
5. Pilih "d":
   - Path = ["a", "d"]
   - "d" adalah node akhir, simpan path ini.

Hasil semua jalur dari "a" ke "d":
```
[
  ['a', 'b', 'c', 'd'],
  ['a', 'b', 'e', 'c', 'd'],
  ['a', 'c', 'd'],
  ['a', 'c', 'e', 'd'],
  ['a', 'd']
]
```

#### 2. Fungsi `terpendekPath`

Fungsi ini menemukan jalur terpendek dari node awal ke node akhir dalam graf.

**Proses:**
- Mulai dari node awal.
- Tambahkan node awal ke path saat ini.
- Jika node awal adalah node akhir, kembalikan path.
- Jika node awal tidak ada di graf, kembalikan None.
- Untuk setiap tetangga node awal, jika tetangga belum ada di path saat ini, rekursi untuk menemukan jalur terpendek dari tetangga ke node akhir.
- Simpan path baru jika lebih pendek dari jalur terpendek yang ditemukan sejauh ini.

**Contoh:**
Untuk mencari jalur terpendek dari "a" ke "d":

1. Mulai dari "a", path = ["a"]
2. Tetangga "a": ["b", "c", "d"]
3. Pilih "b":
   - Path = ["a", "b"]
   - Tetangga "b": ["a", "c", "e"]
   - Pilih "c":
     - Path = ["a", "b", "c"]
     - Tetangga "c": ["a", "b", "d", "e"]
     - Pilih "d":
       - Path = ["a", "b", "c", "d"]
       - "d" adalah node akhir, simpan path ini.
   - Pilih "e":
     - Path = ["a", "b", "e"]
     - Tetangga "e": ["b", "c", "d"]
     - Pilih "c":
       - Path = ["a", "b", "e", "c"]
       - Tetangga "c": ["a", "b", "d", "e"]
       - Pilih "d":
         - Path = ["a", "b", "e", "c", "d"]
         - "d" adalah node akhir, simpan path ini.
4. Pilih "c":
   - Path = ["a", "c"]
   - Tetangga "c": ["a", "b", "d", "e"]
   - Pilih "d":
     - Path = ["a", "c", "d"]
     - "d" adalah node akhir, simpan path ini (lebih pendek dari sebelumnya).
5. Pilih "d":
   - Path = ["a", "d"]
   - "d" adalah node akhir, simpan path ini (ini jalur terpendek).

Hasil jalur terpendek dari "a" ke "d":
```
['a', 'd']
```

#### 3. Fungsi `terpanjangPath`

Fungsi ini menemukan jalur terpanjang dari node awal ke node akhir dalam graf.

**Proses:**
- Mulai dari node awal.
- Tambahkan node awal ke path saat ini.
- Jika node awal adalah node akhir, kembalikan path.
- Jika node awal tidak ada di graf, kembalikan None.
- Untuk setiap tetangga node awal, jika tetangga belum ada di path saat ini, rekursi untuk menemukan jalur terpanjang dari tetangga ke node akhir.
- Simpan path baru jika lebih panjang dari jalur terpanjang yang ditemukan sejauh ini.

**Contoh:**
Untuk mencari jalur terpanjang dari "a" ke "d":

1. Mulai dari "a", path = ["a"]
2. Tetangga "a": ["b", "c", "d"]
3. Pilih "b":
   - Path = ["a", "b"]
   - Tetangga "b": ["a", "c", "e"]
   - Pilih "c":
     - Path = ["a", "b", "c"]
     - Tetangga "c": ["a", "b", "d", "e"]
     - Pilih "d":
       - Path = ["a", "b", "c", "d"]
       - "d" adalah node akhir, simpan path ini.
   - Pilih "e":
     - Path = ["a", "b", "e"]
     - Tetangga "e": ["b", "c", "d"]
     - Pilih "c":
       - Path = ["a", "b", "e", "c"]
       - Tetangga "c": ["a", "b", "d", "e"]
       - Pilih "d":
         - Path = ["a", "b", "e", "c", "d"]
         - "d" adalah node akhir, simpan path ini (lebih panjang dari sebelumnya).
4. Pilih "c":
   - Path = ["a", "c"]
   - Tetangga "c": ["a", "b", "d", "e"]
   - Pilih "d":
     - Path = ["a", "c", "d"]
     - "d" adalah node akhir, simpan path ini (tidak lebih panjang dari yang sebelumnya).
5. Pilih "d":
   - Path = ["a", "d"]
   - "d" adalah node akhir, simpan path ini (tidak lebih panjang dari yang sebelumnya).

Hasil jalur terpanjang dari "a" ke "d":
```
[['a', 'b', 'e', 'c', 'd']]
```

Dengan penjelasan ini, Anda bisa melihat bagaimana setiap fungsi bekerja untuk menemukan jalur-jalur dalam graf, baik itu semua jalur, jalur terpendek, atau jalur terpanjang.
````
