# Linked List

## Node

```python
class Node:
    def __init__(self, item):
        self.data = item
        self.next = None
    # get_set
    def getData(self):
        return self.data
    def getNext(self):
        return self.next
    def setData(self, item):
        self.data = item
    def setNext(self, pt):
        self.next = pt
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def isEmpty(self):
        return self.head is None

    def addNode(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp

    def addFront(self, item):
        self.addNode(item)

    def size(self):
        current = self.head
        count = 0
        while current is not None:
            count += 1
            current = current.getNext()
        return count

    def displayList(self):
        current = self.head
        temp = '['
        while current is not None:
            temp += str(current.getData()) + ' '
            current = current.getNext()
        temp = temp.strip() + ']'
        return temp

    def search(self, data):
        current = self.head
        while current is not None:
            if data == current.getData():
                return True
            current = current.getNext()
        return False

    def addRear(self, newData):
        temp = Node(newData)
        if self.head is None:
            self.head = temp
        else:
            current = self.head
            while current.getNext() is not None:
                current = current.getNext()
            current.setNext(temp)

    def insertBefore(self, data, target):
        current = self.head
        previous = None
        found = False
        newNode = Node(data) 
        while current is not None and not found:
            if current.getData() == target:
                found = True
            else: 
                previous = current 
                current = current.getNext()
        if found:
            newNode.setNext(current)
            if previous is None:
                self.head = newNode
            else:
                previous.setNext(newNode)

    def insertAfter(self, data, target):
        current = self.head
        found = False
        while current is not None and not found:
            if current.getData() == target:
                found = True
            else:
                current = current.getNext()
        if found:
            temp = Node(data)
            temp.setNext(current.getNext())
            current.setNext(temp)

    def removeFront(self):
        if self.head is None:
            return "List kosong"
        else:
            self.head = self.head.getNext()

    def removeRear(self):
        if self.head is None:
            return "List kosong"
        elif self.head.getNext() is None:
            self.head = None
        else:
            current = self.head
            while current.getNext().getNext() is not None:
                current = current.getNext()
            current.setNext(None)

    def remove(self, data):
        current = self.head
        previous = None
        found = False
        while current is not None and not found:
            if current.getData() == data:
                found = True
            else:
                previous = current
                current = current.getNext()
        if found:
            if previous is None:
                self.head = current.getNext()
            else:
                previous.setNext(current.getNext())
    def concatenate(self, list2):
        if list2.head is not None:
            if self.head is None:
                self.head = list2.head
            else:
                current = self.head
                while current.getNext() is not None:
                    current = current.getNext()
                current.setNext(list2.head)
```

```python
# Membuat instance dari LinkedList
ll = LinkedList()

# Menambahkan elemen di depan dan belakang daftar
ll.addFront(10)
ll.addRear(20)
ll.addRear(30)

# Menampilkan daftar
print(ll.displayList())  # Output: [10 20 30]

# Mencari elemen dalam daftar
print(ll.search(20))  # Output: True
print(ll.search(40))  # Output: False

# Menyisipkan elemen sebelum dan sesudah elemen tertentu
ll.insertBefore(15, 20)
ll.insertAfter(25, 20)

# Menampilkan daftar setelah penyisipan
print(ll.displayList())  # Output: [10 15 20 25 30]

# Menghapus elemen dari depan dan belakang
ll.removeFront()
ll.removeRear()

# Menampilkan daftar setelah penghapusan
print(ll.displayList())  # Output: [15 20 25]

# Menghapus elemen tertentu
ll.remove(20)

# Menampilkan daftar setelah penghapusan elemen tertentu
print(ll.displayList())  # Output: [15 25]
```

```python
# Membuat instance dari LinkedList
li1 = LinkedList()
li2 = LinkedList()

# Menambahkan elemen ke li1
li1.addRear(10)
li1.addRear(20)
li1.addRear(30)

# Menambahkan elemen ke li2
li2.addRear(40)
li2.addRear(50)
li2.addRear(60)

# Menampilkan kedua daftar sebelum penggabungan
print("List 1 sebelum penggabungan:", li1.displayList())  # Output: [10 20 30]
print("List 2 sebelum penggabungan:", li2.displayList())  # Output: [40 50 60]

# Menggabungkan kedua daftar
li1.concatenate(li2)

# Menampilkan daftar setelah penggabungan
print("List setelah penggabungan:", li1.displayList())  # Output: [10 20 30 40 50 60]
```

```python
class LinkedList:
    def __init__(self):  # Konstruktor untuk inisialisasi linked list
        self.head = None  # Inisialisasi kepala linked list sebagai None

    def isEmpty(self):  # Metode untuk memeriksa apakah linked list kosong
        return self.head is None  # Mengembalikan True jika kepala adalah None, sebaliknya False

    def addNode(self, item):  # Metode untuk menambahkan node di depan linked list
        temp = Node(item)  # Membuat node baru dengan data 'item'
        temp.setNext(self.head)  # Mengatur node baru untuk menunjuk ke kepala saat ini
        self.head = temp  # Menetapkan kepala ke node baru, sehingga node baru menjadi kepala baru

    def addFront(self, item):  # Metode untuk menambahkan node di depan linked list
        self.addNode(item)  # Memanggil metode addNode karena ini hanya operasi yang sama

    def size(self):  # Metode untuk menghitung jumlah node dalam linked list
        current = self.head  # Pointer saat ini mengarah ke kepala linked list
        count = 0  # Inisialisasi penghitung jumlah node
        while current is not None:  # Iterasi melalui linked list sampai pointer saat ini adalah None (akhir linked list)
            count += 1  # Tambahkan 1 ke penghitung
            current = current.getNext()  # Pindah ke node berikutnya
        return count  # Mengembalikan jumlah node dalam linked list

    def displayList(self):  # Metode untuk menampilkan isi linked list
        current = self.head  # Pointer saat ini mengarah ke kepala linked list
        temp = '['  # Inisialisasi string untuk menyimpan data linked list
        while current is not None:  # Iterasi melalui linked list sampai pointer saat ini adalah None (akhir linked list)
            temp += str(current.getData()) + ' '  # Tambahkan data dari node saat ini ke string
            current = current.getNext()  # Pindah ke node berikutnya
        temp = temp.strip() + ']'  # Hapus spasi ekstra dan tambahkan tanda kurung siku tutup
        return temp  # Mengembalikan string yang berisi data linked list

    def search(self, data):  # Metode untuk mencari data dalam linked list
        current = self.head  # Pointer saat ini mengarah ke kepala linked list
        while current is not None:  # Iterasi melalui linked list sampai pointer saat ini adalah None (akhir linked list)
            if data == current.getData():  # Jika data ditemukan di node saat ini
                return True  # Mengembalikan True
            current = current.getNext()  # Pindah ke node berikutnya
        return False  # Mengembalikan False jika data tidak ditemukan

    def addRear(self, newData):  # Metode untuk menambahkan node di belakang linked list
        temp = Node(newData)  # Membuat node baru dengan data 'newData'
        if self.head is None:  # Jika linked list kosong
            self.head = temp  # Atur kepala menjadi node baru
        else:  # Jika linked list tidak kosong
            current = self.head  # Pointer saat ini mengarah ke kepala linked list
            while current.getNext() is not None:  # Iterasi melalui linked list sampai menemukan node terakhir
                current = current.getNext()  # Pindah ke node berikutnya
            current.setNext(temp)  # Menambahkan node baru setelah node terakhir

    def insertBefore(self, data, target):  # Metode untuk menyisipkan node sebelum node tertentu dalam linked list
        current = self.head  # Pointer saat ini mengarah ke kepala linked list
        previous = None  # Pointer sebelumnya diatur ke None
        found = False  # Inisialisasi penanda apakah node target ditemukan
        newNode = Node(data)  # Membuat node baru dengan data 'data'
        while current is not None and not found:  # Iterasi melalui linked list sampai menemukan node target atau mencapai akhir linked list
            if current.getData() == target:  # Jika data node saat ini sama dengan target
                found = True  # Atur penanda menjadi True
            else:
                previous = current  # Simpan node saat ini sebagai node sebelumnya
                current = current.getNext()  # Pindah ke node berikutnya
        if found:  # Jika node target ditemukan
            newNode.setNext(current)  # Node baru menunjuk ke node saat ini
            if previous is None:  # Jika node target adalah kepala
                self.head = newNode  # Node baru menjadi kepala
            else:
                previous.setNext(newNode)  # Node sebelumnya menunjuk ke node baru

    def insertAfter(self, data, target):  # Metode untuk menyisipkan node setelah node tertentu dalam linked list
        current = self.head  # Pointer saat ini mengarah ke kepala linked list
        found = False  # Inisialisasi penanda apakah node target ditemukan
        while current is not None and not found:  # Iterasi melalui linked list sampai menemukan node target atau mencapai akhir linked list
            if current.getData() == target:  # Jika data node saat ini sama dengan target
                found = True  # Atur penanda menjadi True
            else:
                current = current.getNext()  # Pindah ke node berikutnya
        if found:  # Jika node target ditemukan
            temp = Node(data)  # Membuat node baru dengan data 'data'
            temp.setNext(current.getNext())  # Node baru menunjuk ke node setelah node saat ini
            current.setNext(temp)  # Node saat ini menunjuk ke node baru

    def removeFront(self):  # Metode untuk menghapus node di depan linked list
        if self.head is None:  # Jika linked list kosong
            return "List kosong"  # Mengembalikan pesan bahwa linked list kosong
        else:
            self.head = self.head.getNext()  # Atur kepala untuk menunjuk ke node berikutnya, sehingga menghapus node pertama

    def removeRear(self):  # Metode untuk menghapus node di belakang linked list
        if self.head is None:  # Jika linked list kosong
            return "List kosong"  # Mengembalikan pesan bahwa linked list kosong
        elif self.head.getNext() is None:  # Jika hanya ada satu node dalam linked list
            self.head = None  # Atur kepala menjadi None, sehingga linked list kosong
       
    def concatenate(self, list2):  # Metode untuk menggabungkan dua linked list
    if list2.head is not None:  # Memeriksa apakah linked list kedua tidak kosong
        if self.head is None:  # Jika linked list pertama kosong
            self.head = list2.head  # Atur kepala linked list pertama ke kepala linked list kedua
        else:  # Jika linked list pertama tidak kosong
            current = self.head  # Pointer saat ini mengarah ke kepala linked list pertama
            while current.getNext() is not None:  # Iterasi melalui linked list pertama sampai mencapai node terakhir
                current = current.getNext()  # Pindah ke node berikutnya
            current.setNext(list2.head)  # Menetapkan node terakhir dari linked list pertama untuk menunjuk ke kepala linked list kedua
```
